# Data Import

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

car_data = pd.read_csv('/content/sample_data/car_sales_data.csv')
display(car_data.head())
print(car_data.shape)
```

# Data Vizualization

```python
plt.figure(figsize=(10,6))
plt.hist(car_data['Price'], bins=20, color='skyblue',edgecolor='black')
plt.title('Distribution of Car prices')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.hist(car_data['Engine size'], bins=15, color='orchid', edgecolor='black')
plt.title('Distribution of engine size')
plt.xlabel('Engine size')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(14,8))
plt.scatter(car_data.index ,car_data['Year of manufacture'],color='crimson', alpha=0.5, s=10)
plt.title('Year of manufacture')
plt.xlabel('indice')
plt.ylabel('Year of manufacture')
plt.grid(True)
plt.show()
```
