About Dataset
This dataset contains coffee shop transaction records, including details about sales, payment type, time of purchase, and customer preferences.
It is specifically curated for data visualization, dashboard building, and business analytics projects in tools like Power BI, Tableau, and Python visualization libraries (Matplotlib, Seaborn, Plotly).

With attributes covering time of day, weekdays, months, coffee types, and revenue, this dataset provides a strong foundation for analyzing customer behavior, sales patterns, and business performance trends.

Dataset Structure-

File format: CSV
Columns (features):

hour_of_day ‚Üí Hour of purchase (0‚Äì23)

cash_type ‚Üí Mode of payment (cash / card)

money ‚Üí Transaction amount (in local currency)

coffee_name ‚Üí Type of coffee purchased (e.g., Latte, Americano, Hot Chocolate)

Time_of_Day ‚Üí Categorized time of purchase (Morning, Afternoon, Night)

Weekday ‚Üí Day of the week (e.g., Mon, Tue, ‚Ä¶)

Month_name ‚Üí Month of purchase (e.g., Jan, Feb, Mar)

Weekdaysort ‚Üí Numeric representation for weekday ordering (1 = Mon, 7 = Sun)

Monthsort ‚Üí Numeric representation for month ordering (1 = Jan, 12 = Dec)

Date ‚Üí Date of transaction (YYYY-MM-DD)

Time ‚Üí Exact time of transaction (HH:MM:SS)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

cofee_sales = pd.read_csv('/content/sample_data/Coffe_sales.csv')
display(cofee_sales.head())
```

```python
plt.figure(figsize=(10,6))
plt.hist(cofee_sales['money'], bins=20, color='skyblue',edgecolor='black')
plt.xlabel('Transaction Amount')
plt.ylabel('Frecuency')
plt.title('Distribution of transaction')
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.hist(cofee_sales['Weekdaysort'], bins=20, color='skyblue', edgecolor='black')
plt.xlabel('Weekday sort')
plt.ylabel('Frecuency')
plt.title('Distribution of Transaction')
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.hist(cofee_sales['Monthsort'], bins=20, color='skyblue', edgecolor='black')
plt.xlabel('Monthsort')
plt.ylabel('Frequency')
plt.title('Distribution of Transaction per month')
plt.show()
```
